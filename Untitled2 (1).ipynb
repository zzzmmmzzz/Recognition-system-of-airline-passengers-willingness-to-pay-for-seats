```python
for i in range (300, train.shape[1], 300):
    model_pca = PCA(n_components=i)
    model_pca = model_pca.fit(train)
    train_pca = model_pca.transform(train)
    components=model_pca.components_
    components_var=model_pca.explained_variance_
    print("-----------------------: ",model_pca.explained_variance_ratio_)
    if sum(model_pca.explained_variance_ratio_) >= 0.9:
        print(sum(model_pca.explained_variance_ratio_))
        break;

print("n_components  ",i)
print(train_pca)
print(components)
print(components_var)
print(train_pca.shape)
print(components.shape)
print(components_var.shape)
```

```python
import lightgbm as lgb
from sklearn.metrics import roc_auc_score



def computeAuc(y_true, y_score):
    auc = roc_auc_score(y_true, y_score)
    print("auc=",auc)
    return auc


lgb_train = lgb.Dataset(train_pca, y1)
params = {
    'task':'train',
    'boosting_type':'gbdt',
    'objective':'binary',
    'metric':{'auc'},
    'num_leaves':10,
    'learning_rate':0.0001,
    'feature_fraction':0.4,
    'bagging_fraction':0.7,
    'random_state':1024,
    'lambda':0.2,
    'max_depth':4,
    'is_unbalance':True
}
print("----------------------")
gbm = lgb.train(params, lgb_train, num_boost_round=1000)
predicted_probs_train = gbm.predict(train_pca)
predicted_probs_train = [x for x in predicted_probs_train]
computeAuc(y1, predicted_probs_train)
```
